```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
%matplotlib inline
```

```python
filepath="laptop_pricing_dataset_mod2.csv"
df= pd.read_csv(filepath, header=0)
df.head(5)
```

Regression Plot

```python
sns.regplot(x="CPU_frequency", y="Price", data=df)
plt.ylim(0,)
```

```python
sns.regplot(x="Screen_Size_inch", y="Price", data=df)
plt.ylim(0,)
```

```python
sns.regplot(x="Weight_pounds", y="Price", data=df)
plt.ylim(0,)
```

```python
for param in ["CPU_frequency","Screen_Size_inch","Weight_pounds"]:
    print(f"Correlation of Price and {param} is ",df[[param,"Price"]].corr())
```

Descriptive Statical Analysis

```python
print(df.describe())
print(df.describe(include=['object']))
```

Group BY

```python
df_gptest=df[['GPU','CPU_core','Price']]
grouped_test1=df_gptest.groupby(['GPU','CPU_core','Price'],as_index=False).mean()
print(grouped_test1)
```

```python
# Check for duplicate combinations of values in the pivot columns
duplicate_combinations = df.duplicated(subset=['GPU', 'CPU_core'], keep='first')
duplicate_combinations_rows = df[duplicate_combinations]
print("Duplicate Combination Rows:")
print(duplicate_combinations_rows)

# Perform the pivot operation
try:
    grouped_pivot = df.pivot(index='GPU', columns='CPU_core')
    print(grouped_pivot)
except ValueError as e:
    print("Error:", e)
```

Pearson Correlation and P-Values

```python
for param in ['RAM_GB','CPU_frequency','Storage_GB_SSD','Screen_Size_inch','Weight_pounds','CPU_core','OS','GPU','Category']:
    p_coef, p_value= stats.pearsonr(df[param],df['Price'])
    print(param)
    print("The Pearson Correlation Coefficient for ",param," is ",p_coef," with a P-value of P= ",p_value)
```
